## Performing logistic regression with 80% train and 20%test dataset

```python
model=LogisticRegression()
```

```python
Y=df1.iloc[:,-1]
X=df1.iloc[:,:-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2) ## Splitting the train model and test model, so 80% is trained model and test model is 20%
model.fit(x_train, y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
confusion_matrix(y_predict,y_test)
```

## Here the accuracy score is 100% by applying logistic regression

```python
accuracy_score(y_test,y_predict)*100
```

## 210 regression in neural model

```python
model=MLPClassifier(hidden_layer_sizes=(150,60))
model.fit(x_train, y_train)
```

```python
y_predict=model.predict(x_test)
confusion_matrix(y_predict,y_test)
```

## finally i also got the accuracy score is 100%

```python
accuracy_score(y_test,y_predict)*100
```

# Random forest

```python
clf = RandomForestClassifier(n_estimators=46,random_state=8)
clf.fit(x_train, y_train) #Training the model

predicted_classes = clf.predict(x_test)
accuracy = accuracy_score(y_test,predicted_classes) #evaluating the test data
print("random forest")
print(accuracy*100)
```
